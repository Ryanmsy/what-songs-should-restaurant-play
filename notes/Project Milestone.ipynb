# Restaurant Song Recommendation Engine — Project Roadmap

## Overview
This project builds a recommendation engine that suggests what songs a restaurant should play, based on the restaurant's vibe and business attributes.

Two datasets power the engine:
- **Spotify** (`tracks_features.csv`) 
- **Yelp** (`yelp_clean.csv`) 

The end goal is to match a restaurant's identity (from Yelp) to songs that fit its atmosphere (from Spotify)

---

## Milestone 1 — Redundancy Analysis & Dimensionality Reduction

Perform Covariance matrix / correlation

covariance matrix = compares the variance of each variable vs the covariance (does x and y variable move in the same direction)

the number of columns = number of rows because it is a square matrix, so if we have 5 features, we have 5 rows and 5 columns

the diagonal elements are the variances of each variable (x,y,z)

### Steps

1. Remove all categorical data so that the dataset $X$ consists of only continuous variables
2. Standardize features (z-score) so scale differences don't distort results
3. Compute the covariance matrix
4. Visualize as a heatmap Spotify correlation map
5. Visualize the subspace — scatter plots showing redundant vs independent feature pairs
6. Visualize as a heatmap — Yelp correlation map
7. **PCA** — find which columns are the most important
8. **SVD** — singular value decomposition
9. **Eigenvalues & Eigenvectors** — the math behind PCA
